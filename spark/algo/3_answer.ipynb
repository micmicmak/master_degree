### Question 1

```python
numPartitions = 10
```

```python
lines = sc.textFile('./data/adj_noun_pairs.txt', numPartitions)
pairs = lines.map(lambda l: tuple(l.split())).filter(lambda p: len(p)==2)
pairs.cache()
pairs.collect()
```

```python
# FILL IN YOUR CODE HERE
def f(item):
    answer = 'No such pair'
    for p in item:
        if p[1] == 'unification':
            answer = p
            break
    yield answer
lst = pairs.mapPartitions(f).collect()
lst
```

```python
final_result = None
for i in lst:
    if i != 'No such pair':
        final_result = i
        break
final_result
```

### Question 2

```python
x = 'abcccbcbcacaccacaabb'
y = 'abcccbcccacaccacaabb'

numPartitions = 4
rdd = sc.parallelize(zip(x,y), numPartitions)
```

```python
# FILL IN YOUR CODE HERE
def f(item):
    tmp = 0
    for i in item:
        if i[1] < i[0]:
            tmp += 1
        elif i[1] > i[0]:
            tmp -= 1
        else:
            pass
    yield tmp
result = rdd.mapPartitions(f).collect()
sign = None
if sum(result) < 0:
    sign = '<'
elif sum(result) > 0:
    sign = '>'
else:
    sign = '='
print('x {} y'.format(sign))
```
